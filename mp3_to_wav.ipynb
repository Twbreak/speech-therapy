```python
import os
from glob import glob
from tqdm import tqdm  # 進度條套件
import subprocess
```

mp3_to_wav 函式

```python
def mp3_to_wav(mp3_file_name: str, wav_file_name: str):
    ffmpegEXE_path = "ffmpeg-7.0.1-full_build\\bin\\ffmpeg.exe"  # 一定要有此檔案才能執行轉換功能 
    subprocess.call([ffmpegEXE_path, '-i', mp3_file_name, wav_file_name])
```

資料前處理

男生聲音

```python
male_mp3_files = glob("voice_source\\male_mp3\\*")
print(len(male_mp3_files))
male_mp3_files
```

```python
dir_path = "voice_source\\male_wav"
print(f"'{dir_path}' exiet? {os.path.exists(dir_path)}")
if not os.path.exists(dir_path):
    os.mkdir(dir_path)
    print(f"make '{dir_path}'")
```

```python
for mp3_file in tqdm(male_mp3_files):
    wav_file = mp3_file.replace("_mp3", "_wav").replace(".mp3", ".wav")
    mp3_to_wav(mp3_file_name=mp3_file, wav_file_name=wav_file)
```
