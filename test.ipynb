```python
model = whisper.load_model("large")
# 使用Whisper進行語音識別
result = model.transcribe("papaya_1.wav")
result['segments']
```

```python
transcript = result["text"]

# 將識別結果進行文本修正
corrector = T5Corrector()
corrected_transcript = corrector.correct(transcript)

corrected_transcrip_list = list(corrected_transcript['target'])
```
